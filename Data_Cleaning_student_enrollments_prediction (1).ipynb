```python
import re

def clean_purpose(text):

    # Convert to string (handle nan)
    text = str(text)

    # Lowercase
    text = text.lower()

    # Replace s/w or s\w with software
    text = re.sub(r's[/\\]w', 'software', text)

    # Replace slash and backslash with ' and '
    text = re.sub(r'[/\\]', ' and ', text)

    # Replace + and & with ' and '
    text = re.sub(r'[+&]', ' and ', text)

    # Remove brackets and special characters except letters and spaces
    text = re.sub(r'[^a-z\s]', ' ', text)

    # Fix common spelling mistakes
    corrections = {
        'intership': 'internship',
        'internnship': 'internship',
        'internishp': 'internship',
        'intenship': 'internship',
        'placment': 'placement',
        'regrding': 'regarding',
        'enqiry': 'enquiry',
        'enquri': 'enquiry',
        'enqurie': 'enquiry',
        'coustomer': 'customer',
        'salse': 'sales',
        'telesalse': 'telesales',
        'markrting': 'marketing',
        'testeing': 'testing',
        'cyberv': 'cyber',
        'concelling': 'counselling'
    }

    for wrong, correct in corrections.items():
        text = text.replace(wrong, correct)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text


# Apply to dataframe column
df['Purpose'] = df['Purpose'].apply(clean_purpose)
```

```python
import re

def clean_placement(text):

    # Convert to string
    text = str(text)

    # Lowercase
    text = text.lower()

    # Replace slash and backslash with ' and '
    text = re.sub(r'[/\\]', ' and ', text)

    # Replace + and & with ' and '
    text = re.sub(r'[+&]', ' and ', text)

    # Fix common spelling mistakes
    corrections = {
        'supervisior': 'supervisor',
        'fronted': 'frontend',
        'enquiery': 'enquiry',
        'regrding': 'regarding',
        'intrested': 'interested',
        'placments': 'placement',
        'placements': 'placement'
    }

    for wrong, correct in corrections.items():
        text = text.replace(wrong, correct)

    # Remove special characters (keep only letters and spaces)
    text = re.sub(r'[^a-z\s]', ' ', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text


# Apply to Placement column
df['Placement'] = df['Placement'].apply(clean_placement)
```

```python
import re

def clean_training(text):

    # Convert to string
    text = str(text)

    # Lowercase
    text = text.lower()

    # Replace s/w or s\w with software
    text = re.sub(r's[/\\]w', 'software', text)

    # Replace s.t with software testing
    text = re.sub(r'\bs\.t\b', 'software testing', text)

    # Replace slash and backslash with ' and '
    text = re.sub(r'[/\\]', ' and ', text)

    # Replace + and & with ' and '
    text = re.sub(r'[+&]', ' and ', text)

    # Fix spelling mistakes
    corrections = {
        'markrting': 'marketing',
        'pyhton': 'python',
        'cyberv': 'cyber',
        'enqiry': 'enquiry'
    }

    for wrong, correct in corrections.items():
        text = text.replace(wrong, correct)

    # Remove special characters (keep only letters and spaces)
    text = re.sub(r'[^a-z\s]', ' ', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text


# Apply to Training column
df['Training'] = df['Training'].apply(clean_training)
```
