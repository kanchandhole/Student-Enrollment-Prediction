```python
#Drop unnecessary columns
df = df.drop(['Date', 'Name', 'Contact No'], axis=1)
```

```python
 #Handle missing values (if any)
df = df.dropna()   # simple approach
```

```python
df.head(4)
```

```python
#Clean categorical values
df['Purpose'] = df['Purpose'].str.lower().str.strip()
df['Registration'] = df['Registration'].str.lower().str.strip()
```

```python
#Define X and y
X = df[['Purpose']]          # feature
y = df['Registration']       # target
```

In my dataset, 'Purpose' represents the reason why a student visited or showed interest, which can directly impact whether they will register or not.

Other columns like Name, Contact Number, and Date do not contribute meaningful information for prediction, so I removed them.

Since 'Purpose' was the only meaningful feature available, I used it as the input variable (X) to train my model.

```python
X
```

```python
y
```

```python
#Encoding
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
# Feature encoding
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)
X_encoded = encoder.fit_transform(X)
```

```python
X_encoded
```

The sparse_output parameter controls the format of the output.
If set to True, it returns a sparse matrix which is memory efficient.
If set to False, it returns a dense array which is easier to use for further processing like scaling

```python
## Target encoding
le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
y_encoded
```

```python
#Train-Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y_encoded, test_size=0.2, random_state=42
)
```

```python
#Check imbalance
from collections import Counter
print("\nBefore SMOTE:", Counter(y_train))
```

```python
#Apply SMOTE
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)

print("After SMOTE:", Counter(y_train_res))
```

After defining X and y, I first performed encoding and train-test split. Then I applied SMOTE only on the training data to handle class imbalance without causing data leakage.

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_res = scaler.fit_transform(X_train_res)
X_test = scaler.transform(X_test)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train_res, y_train_res)
y_pred_lr = lr.predict(X_test)
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train_res, y_train_res)
y_pred_dt = dt.predict(X_test)
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_res, y_train_res)
y_pred_rf = rf.predict(X_test)
```

```python
#EVALUATION
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score, classification_report
```

```python
print("Precision:", precision_score(y_test, y_pred_dt))
print("Recall:", recall_score(y_test, y_pred_dt))
print("F1 Score:", f1_score(y_test, y_pred_dt))
print("Accuracy:", accuracy_score(y_test, y_pred_dt))
```

```python
print("\n Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_lr))
print(classification_report(y_test, y_pred_lr))
```

The model achieved an accuracy of around 91.7%. However, since the dataset is imbalanced, accuracy alone is not reliable.

For class 0, the model performs very well with high precision and recall. But for class 1, which is the minority class, the recall is only 54%, meaning the model fails to correctly identify many actual registered students.

This indicates that the model is biased towards the majority class. Therefore, metrics like precision, recall, and F1-score provide a better understanding of performance.

```python
print(" Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt))
print(classification_report(y_test, y_pred_dt))
```

The Decision Tree model achieved an accuracy of around 92.6%, which is slightly higher than Logistic Regression.

For the majority class, the model performs very well with high precision and recall. For the minority class, precision improved to 70%, meaning predictions are more accurate, but recall is still low at 54%, indicating that the model misses many actual registered students.

Overall, the model shows slight improvement over Logistic Regression, but still struggles with class imbalance.

```python
print(" Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```

The Random Forest model achieved an accuracy of 92.6%, which is similar to the Decision Tree model.

For the majority class, the model performs very well. However, for the minority class, the recall is still low at 54%, meaning the model fails to identify many actual registered students.

Since the dataset has only one feature and is relatively small, Random Forest does not provide significant improvement over Decision Tree
